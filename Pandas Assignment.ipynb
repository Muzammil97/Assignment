### 1. Create a data series with marks of students : 75, 80, 79, 60

### 5. Sort the data based on Marks obtained. Fill all the 'na' cells with 0

```python
clean = var.fillna(0)


clean = var.replace(np.nan , 0)



sort = clean.sort_values(by=["Total (100)"], ascending=False)


display(sort.head(10))
sort.head(6)[["Student Name","Grade"]]
```

### 9. Display only those students who got failed in examination

```python
display(sort.loc[var.Grade.isin(['F']),['Student Name','Grade']])


fail = sort["Grade"]=="F"
display(sort[fail])
```
